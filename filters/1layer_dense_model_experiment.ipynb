```python
import os
import numpy as np

from keras.models import load_model
from keras.datasets import cifar10
from keras.utils import to_categorical

import matplotlib.pyplot as plt
import seaborn as sns

# load model
model = load_model('../input/hidden/1layer_dense_model.h5')
model.summary()
```

```python
# plotting the weights to the output layer
layer_name = 'dense_1'

for layer in model.layers:
    if layer.get_config()['name'] == layer_name:
        weights = layer.get_weights()

# not considering bias
weights[0].transpose().shape
```

```python
# plot grid
fig = plt.figure(figsize = (16, 16))
for i in range(10):
    image = weights[0][:,i].reshape(28, 28).astype(np.float)
    ax = fig.add_subplot(4, 4, i+1)
    ax.title.set_text('Target: ' + str(i))
    plt.imshow(image)
plt.show()
```

```python
# plot single image
image_index = 0

image = weights[0][:,image_index].reshape(28, 28).astype(np.float)
plt.title('Target: ' + str(image_index))
plt.imshow(image)
```
