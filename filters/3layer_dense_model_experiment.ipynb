```python
import os
import numpy as np

from keras.models import load_model
from keras.datasets import cifar10
from keras.utils import to_categorical

import matplotlib.pyplot as plt
import seaborn as sns

# load model
model = load_model('../input/hidden/3layer_dense_model.h5')
model.summary()
```

```python
# plotting the weights to the first hidden layer
layer_name = 'dense_1'

for layer in model.layers:
    if layer.get_config()['name'] == layer_name:
        weights = layer.get_weights()

# not considering bias
weights[0].transpose().shape
```

```python
# plot grid
fig = plt.figure(figsize = (128, 128))
for i in range(256):
    image = weights[0][:,i].reshape(28, 28).astype(np.float)
    fig.add_subplot(16, 16, i+1)
    plt.imshow(image)
plt.show()
```

```python
# plot single image
image_index = 116
plt.imshow(weights[0][:,image_index].reshape(28, 28).astype(np.float))
```
